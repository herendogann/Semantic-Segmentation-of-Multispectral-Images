# Maximum Likelihood Estimate (MLE) 

**1. You have a coin that you think is biased, you flip it 5 times and get the sequence HTHTH. What is the maximum likelihood estimate for the probability of getting heads?**


The likelihood function ($L$) for a sequence of independent Bernoulli trials (like coin flips) is given by the product of the probabilities of the observed outcomes. In this case, the probability of getting heads ($P$(Heads)) is the parameter we want to estimate. The likelihood function is given by:

$$
L(p) = p^k(1-p)^{n-k}
$$
 
where:

- $p$ is the probability of getting heads,

- $k$ is the number of heads observed,

- $n$ is the total number of coin flips.
 
 


 
In this case, k=3 (three heads), and n=5 (five coin flips).

$$
L(p) = p^3(1-p)^2
$$

To find the MLE, we must find the maximum value for this equation; so we will take the derivative of the likelihood function with respect to $p$ and set it equal to zero:

$$
L'(p) = 3p^2(1-p)^2 + p^3(-2)(1-p) = 0
$$


The critical points for this equation are: [$p$ = 1 , $p$ = 0 , $p$ = 2/3]

We will take the second derivative to find the local maximum and minimum values:
$$
L'(p) = 5p^4 - 8p^3 + 3p^2
$$
$$
L''(P)= 20p^3 - 24p^2 + 6p
$$
 
Now if we evaluate the second derivative for critical points we found:
$$ 
L''(1)= 2
$$
$$
L''(0)= 0  
$$
$$
L''(2/3) = -0.7407
$$ 

Here we find the second derivative we found at ($p$ = 1) is positive and at ($p$ = 2/3) is negative, so ($p$ = 1) indicates a local minimum and  ($p$ = 2/3) indicates the local maximum value. 
Since there is no other local maximum value, we can say that the MLE for the probability of getting heads is:
$$
p = 2/3.
$$

**2. You know that candy prices are normally distributed with mean µ and standard deviation σ. You have three candy pricing 2, 3, 5 lira. What is the maximum likelihood
for µ?**



In a normal distribution, the likelihood function is given by the probability density function (PDF) of the distribution. For a normal distribution with mean μ and standard deviation σ, the PDF is given by:

$$
f(x | μ, \sigma) = \frac{1}{{\sigma\sqrt{2\pi}}} e^{-\frac{1}{2} \left(\frac{x - μ}{\sigma}\right)^2}
$$

The likelihood function is the product of the probabilities for each observation, given the parameters μ and σ. In this case, you have three candy prices: 2, 3, and 5 lira. Let's denote these prices as $x_1$,$x_2$ and $x_3$.

The likelihood function $L(μ)$ is given by:

$$
L(μ, \sigma) = f(x_1 | μ, \sigma) \cdot f(x_2| μ, \sigma) \cdot f(x_3 | μ, \sigma)
$$
$$
L(μ, \sigma) = f(2 | μ, \sigma) \cdot f(3 | μ, \sigma) \cdot f(5 | μ, \sigma)
$$

Taking the $\ln$ of the likelihood function gives the log-likelihood function:

$$
\ln(L(μ, \sigma)) = -\frac{3}{2} \ln(2\pi) - 3 \ln(\sigma) - \frac{1}{2\sigma^2} \left((2 - μ)^2 + (3 - μ)^2 + (5 - μ)^2\right)
$$

We are trying to find the value of μ that maximizes the log-likelihood function. So, we take the derivative of the log-likelihood function with respect to μ, set it equal to zero, and solve for μ:

$$
\frac{d}{dμ} \ln(L(μ, \sigma)) = 0
$$
$$
(2 - μ) + (3 - μ) + (5 - μ) = 0
$$
$$
0 = 10 - 3μ
$$
$$
3μ = 10
$$

$$
μ = 10/3
$$

So, MLE for μ is $10/3$ or approximately $ 3.33 $ lira.

**4. Suppose that X1, ..., Xn form a random sample from a uniform distribution on the interval (0, θ), where of the parameter θ > 0 but is unknown. Please find MLE of θ.**




The probability density function (pdf) of a uniform distribution on the interval (0, $\theta$) is given by:

$$
f(x; \theta) = \frac{1}{\theta} \text{ for } 0 < x < \theta
$$

The likelihood function is the product of the individual probability densities for each observation in the sample:

$$
L(\theta) = \prod_{i=1}^{n} f(x_i; \theta)
$$

$$
L(\theta) = \frac{1}{\theta^n} \prod_{i=1}^{n} \mathbb{I}(0 < x_i < \theta)
$$


The log-likelihood is the natural logarithm of the likelihood function:

$$
\ell(\theta) = \ln(L(\theta))
$$

$$ 
\ell(\theta) = -n \ln(\theta) + \ln\left(\prod_{i=1}^{n} \mathbb{I}(0 < x_i < \theta)\right)
$$

Taking the derivative of the log-likelihood with respect to $\theta$ and setting it equal to zero:

$$
-\frac{n}{\theta} + \frac{1}{\theta} \sum_{i=1}^{n} \mathbb{I}(0 < x_i < \theta) = 0 
$$

Solving for $\theta$, we get:

$$
\frac{1}{\theta} \sum_{i=1}^{n} \mathbb{I}(0 < x_i < \theta) = \frac{n}{\theta}
$$

$$
 \sum_{i=1}^{n} \mathbb{I}(0 < x_i < \theta) = n
$$

The solution to this equation is the MLE for $\theta$, which is the maximum observed value in the sample:

$$
 \hat{\theta}_{\text{MLE}} = \max(X_1, X_2, \ldots, X_n) 
$$

```python
def gaussian_naive_bayes(x_train, y_train, x_test, y_test):
    classes, params = fit_gaussian_nb(x_train, y_train)
    predictions = predict_classes(x_test, classes, params)
    accuracy = accuracy_score(y_test, predictions)
    return accuracy

def fit_gaussian_nb(X, y):
    parameters = {}
    for c in np.unique(y):
        X_c = X[y == c]
        parameters[c] = {"mean": X_c.mean(axis=0),"variance": X_c.var(axis=0), "prior": X_c.shape[0] / X.shape[0]}
    return np.unique(y), parameters

def calculate_probability(X, mean, var):
    return (-0.5 * np.sum(np.log(2. * np.pi * var))) - (0.5 * np.sum(((X - mean) ** 2) / (2 * var), 1)) 

def predict_probability(X, classes, parameters):
    probabilites = np.zeros((X.shape[0], len(classes)))
    for c in classes:
        p = parameters[c]
        probabilites[:, c] = calculate_probability(X, p["mean"], p["variance"]) + np.log(p["prior"])
    return probabilites

def predict_classes(X, classes, parameters):
    return classes[np.argmax(predict_probability(X, classes, parameters), axis=1)]


def accuracy_score(labels_test, predictions):
    count = 0
    for i in range(len(labels_test)):
        if labels_test[i] == predictions[i]:
            count += 1

    return count / len(labels_test)
```

```python
from sklearn.model_selection import train_test_split

test_channels = {0: 'red', 1: 'green', 2: 'blue', 3: 'infrared1', 4: 'infrared2', 5: 'infrared3', 6: 'rgb', 7: 'infrared', 8: 'all'}
train_data = train_data[train_mask != 0]
train_labels = train_labels[train_mask != 0]

X_train, X_test, y_train, y_test = train_test_split(train_data, train_labels,test_size=0.2)

accuracies = {}

for i in test_channels.keys():
    if i < 6:
        accuracies[test_channels[i]] = gaussian_naive_bayes(X_train[:,i:i+1], y_train, X_test[:,i:i+1], y_test)
    elif i == 6:
        accuracies[test_channels[i]] = gaussian_naive_bayes(X_train[:,0:3], y_train, X_test[:,0:3], y_test)
    elif i == 7:
        accuracies[test_channels[i]] = gaussian_naive_bayes(X_train[:,3:6], y_train, X_test[:,3:6], y_test)
    elif i == 8:
        accuracies[test_channels[i]] = gaussian_naive_bayes(X_train, y_train, X_test, y_test)

for key , val in accuracies.items():
    print(f"Accuracy for {key} channel(s):    \t{(val * 100.0):.2f}%")
```
